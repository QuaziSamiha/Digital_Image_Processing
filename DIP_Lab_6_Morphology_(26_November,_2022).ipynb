```python
fig1_url = '/content/drive/MyDrive/Colab Notebooks/images/morphology1.jpg'
fig1 = cv2.resize(cv2.imread(fig1_url), (250,250))
# cv2_imshow(fig1)
```

```python
fig2_url = '/content/drive/MyDrive/Colab Notebooks/images/morphology2.jpg'
fig2 = cv2.resize(cv2.imread(fig2_url), (250,250))
# cv2_imshow(fig2)
```

```python
fig3_url = '/content/drive/MyDrive/Colab Notebooks/images/morphology3.jpg'
fig3 = cv2.resize(cv2.imread(fig3_url), (300,200))
# cv2_imshow(fig3)
```

```python
fig4_url = '/content/drive/MyDrive/Colab Notebooks/images/morphology4.jpg'
fig4 = cv2.resize(cv2.imread(fig4_url), (350,200))
# cv2_imshow(fig4)
```

```python
fig5_url = '/content/drive/MyDrive/Colab Notebooks/images/morphology5.jpg'
fig5 = cv2.resize(cv2.imread(fig5_url), (200,250))
# cv2_imshow(fig5)
```

```python
# Thresholding is a simple and efficient technique to perform basic segmentation in an image, and to binarize it (turn it into a binary image) where pixels are either 0 or 1
ret,th1 = cv2.threshold(fig1,125,125,cv2.THRESH_BINARY)
# np.ones() is a method that returns an array of given shape and type filled with ones
kernel=np.ones((4,4), np.uint8) # syntax: np.ones(shape, dtype)
erosion=cv2.erode(th1,kernel,iterations=6)

# cv2_imshow(th1)
cv2_imshow(fig1)
cv2_imshow(erosion)
```

```python
ret,th1 = cv2.threshold(fig2,125,125,cv2.THRESH_BINARY)
kernel=np.ones((1,2), np.uint8)
erosion=cv2.erode(th1,kernel,iterations=2)
dilation=cv2.dilate(erosion,kernel,iterations=2)

cv2_imshow(fig2)
cv2_imshow(dilation)
```

```python
fig3_gray = cv2.cvtColor(fig3, cv2.COLOR_BGR2GRAY)
ret,th1 = cv2.threshold(fig3_gray,125,125,cv2.THRESH_BINARY)
kernel=np.ones((3,3), np.uint8)
erosion=cv2.erode(th1,kernel,iterations=2)
dilation=cv2.dilate(th1,kernel,iterations=2)
gradient = cv2.morphologyEx(th1, cv2.MORPH_GRADIENT, kernel)

cv2_imshow(fig3_gray)
cv2_imshow(erosion)
cv2_imshow(dilation)
cv2_imshow(gradient)
```

```python
ret,th1 = cv2.threshold(fig4,125,125,cv2.THRESH_BINARY)
kernel=np.ones((3,3), np.uint8)
opening=cv2.morphologyEx(th1, cv2.MORPH_OPEN, kernel)

# cv2_imshow(opening)
top_hat_transform = cv2.subtract(fig4, opening)
cv2_imshow(fig4)
cv2_imshow(top_hat_transform)
```

```python
ret,th1 = cv2.threshold(fig4,125,125,cv2.THRESH_BINARY)
kernel=np.ones((3,3), np.uint8)
closing=cv2.morphologyEx(th1, cv2.MORPH_CLOSE, kernel)

# cv2_imshow(opening)
bottom_hat_transform = cv2.subtract(closing, fig4)
cv2_imshow(fig4)
cv2_imshow(bottom_hat_transform)
```
