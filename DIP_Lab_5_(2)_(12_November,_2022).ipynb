```python
#image padding

img=cv2.imread('/content/drive/MyDrive/Colab Notebooks/images/lina.JPEG')
img=cv2.resize(img,(202,202))
img=img[:,:,0]
for i in range(202):
  img[i][0]=0
  img[i][201]=0
  img[201][i]=0
  img[0][i]=0
cv2_imshow(img)

img2=cv2.imread('/content/drive/MyDrive/Colab Notebooks/images/lina.JPEG')
img2=cv2.resize(img2,(202,202))
img2=img2[:,:,0]
for i in range(202):
  img2[i][0]=0
  img2[i][201]=0
  img2[201][i]=0
  img2[0][i]=0


#cv2_imshow(img2)
```

```python
def mul_and_sum(k,mat):
  sum=0
  for i in range(len(k)):
    for j in range(len(k[0])):
      sum+=k[i][j]*mat[i][j]
  return sum

#prewitt
#horizontal
kernel= [[-1,-1,-1],
          [0,0,0],
          [1,1,1]]

for i in range(1,200):
  for j in range(1,200):
    mat=[[img[i-1][j-1],img[i-1][j],img[i-1][j+1]],
         [img[i][j-1],img[i][j],img[i][j+1]],
         [img[i+1][j-1],img[i+1][j],img[i+1][j+1]]]
    if (mul_and_sum(kernel,mat))>48:
      img2[i][j]=255
    else:
      img2[i][j]=0

cv2_imshow(img2)
#cv2_imshow(img3)
```
